```python
# Simular una ecuacion cuadratica, con valores aleatorios y encontrar la solucion con formula general
# [4,2,4]

a = np.random.randint(1,10)
b = np.random.randint(1,10)
c = np.random.randint(1,10)

print(f"Valores de la ecuación cuadrática: a = {a}, b = {b}, c = {c}")

discriminante = b**2 - 4*a*c
if discriminante >= 0:
    raiz_discriminante = np.sqrt(discriminante)
    x1 = (-b + raiz_discriminante) / (2*a)
    x2 = (-b - raiz_discriminante) / (2*a)
    print(f"Las soluciones de la ecuación cuadrática son: x1 = {x1}, x2 = {x2}")
else:
    print("La ecuación cuadrática no tiene solucion")
```
