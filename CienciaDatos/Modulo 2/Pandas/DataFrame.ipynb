```python
# Ingresar un jugador por teclado y localizarlo en el dataFrame con nombre y apellido

import pandas as pd

# 1. Cargar el archivo CSV
df = pd.read_csv('nba.csv')

# 2. Solicitar el nombre por teclado
nombre_buscado = input("Ingresa el nombre y apellido del jugador: ")

# 3. Localizar al jugador
# Usamos .str.lower() en ambos lados para que la búsqueda sea más flexible
resultado = df[df['Name'].str.lower() == nombre_buscado.lower()]

# 4. Mostrar el resultado
if not resultado.empty:
    print("\n--- Jugador Localizado ---")
    print(resultado)
else:
    print(f"\nNo se encontró a '{nombre_buscado}' en el archivo nba.csv.")
```

# A review of the .apply() method on single columns

```python
b = pd.read_csv("jamesbond.csv", index_col= "Film")
b.sort_index(inplace=True)
b.head(3)
```

```python
# Tengo una columna y esa columna la voy a limpiar de caracteres raros y esa nueva columna la voy a almacenar
```

```python
def convert_to_string_add_millions(number):
    return str(number) + " Millones"
b["Box Office"] = b["Box Office"].apply(convert_to_string_add_millions)
```

```python
b
```

```python
columns = ["Box Office", "Budget", "Bond Actor Salary"]
for col in columns:
    b[col] = b[col].apply(convert_to_string_add_millions)
```

```python
b
```

```python
import pandas as pd
b = pd.read_csv("jamesbond.csv", index_col= "Film")
b.sort_index(inplace=True)
b.head(3)
```

```python
def good_movie(row):
    actor = row["Actor"]
    budget = row["Budget"]

    if actor == "Pierce Brosnan":
        return "The Best"
    elif actor == "Roger Moore" and budget > 40:
        return "Enjoyable"
    else:
        return "I have no clue"
b.apply(good_movie, axis = 1)
```
