## Resumen del DataFrame Chicago.csv

```python
import pandas as pd
```

```python
data = pd.read_csv('../Pandas/chicago.csv')
```

```python
# Elimina símbolos y convierte a numérico
data['Employee Annual Salary'] = data['Employee Annual Salary'].replace('[\$,]', '', regex=True).astype(float)
```

```python
media = data['Employee Annual Salary'].mean()
mediana = data['Employee Annual Salary'].median()
moda = data['Employee Annual Salary'].mode()[0]

print(f'Media: {media}, Mediana: {mediana}, Moda: {moda}')
```

```python
resumen = data.describe()
print(resumen)
```

```python
Varianza = data['Employee Annual Salary'].var()
Desviacion_Estandar = data['Employee Annual Salary'].std()
print(f'Varianza: {Varianza}, Desviación Estándar: {Desviacion_Estandar}')
```

```python
Rango = data['Employee Annual Salary'].max() - data['Employee Annual Salary'].min()
print(f'Rango: {Rango}')
```
