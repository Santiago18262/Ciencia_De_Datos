```python
# CSV

# Importa el módulo nativo de Python para manejar archivos CSV
import csv 
# Importa la librería NumPy para realizar operaciones con arreglos numéricos
import numpy 

# Define la ruta relativa del archivo de datos 
filename = '../Pandas/google_stock_pricecopy.csv' 

# Abre el archivo en modo lectura ('r'); devuelve un objeto de archivo 'raw_data'
raw_data = open(filename, 'r') 

# Crea un objeto 'reader' del [módulo csv de Python]
# delimiter=',': Define la coma como separador de columnas
# quoting=csv.QUOTE_NONE: Indica al lector que [no procese comillas especiales]
reader = csv.reader(raw_data, delimiter= ',', quoting= csv.QUOTE_NONE) 

# Convierte el iterador 'reader' en una lista de Python (donde cada fila es una sublista de strings)
x = list(reader) 

# Crea un [arreglo de NumPy (ndarray) a partir de la lista
# .astype('float'): Convierte todos los elementos de texto a números decimales (float)
data = numpy.array(x).astype('float') 

# Imprime la [forma o dimensiones (shape) del arreglo (ej. número de filas y columnas)
print(data.shape) 
```

```python
# NumPy

# Importa la función loadtxt desde la biblioteca NumPy, usada para cargar archivos de texto plano
from numpy import loadtxt

# Define la ruta del archivo CSV a cargar en una variable para mayor claridad
filename = '../Pandas/google_stock_pricecopy.csv'

# Abre el archivo en modo lectura binaria ('rb'). Esto es útil para loadtxt en algunas plataformas
raw_data = open(filename, 'rb')

# Carga el archivo: lee los datos del file handler, especifica que la coma separa las columnas
data = loadtxt(raw_data, delimiter = ",")

# Muestra la forma (dimensiones) de la matriz NumPy resultante (filas, columnas)
print(data.shape)
```

```python
# Pandas

import pandas as pd # Importa la librería pandas
filename = '../Pandas/nba.csv' # Define una variable con la ruta del archivo CSV a cargar
names = ['Name', 'Team', 'Number', 'Position', 'Age'] # Crea una lista con los nombres que tendrán las columnas
data = pd.read_csv(filename, names = names) # Lee el archivo CSV, asigna los nombres definidos y guarda todo en un DataFrame llamado 'data'
print(data.shape) # Imprime la forma del DataFrame (número de filas, número de columnas)
```

```python
# Pandas

import pandas as pd # Importa la biblioteca pandas
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data' # Define una variable con la URL directa al archivo de datos crudos (CSV).
dataframe = pd.read_csv(url, header = None) # Lee el archivo CSV desde la URL y carga los datos en un DataFrame; 'header = None' indica que la primera fila no contiene nombres de columnas.
print(dataframe.shape) # Imprime la forma del DataFrame (número de filas y columnas)
```
