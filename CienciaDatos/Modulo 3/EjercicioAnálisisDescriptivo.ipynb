```python
import pandas as pd
```

```python
data = pd.read_csv('../Modulo 2/Pandas/fortune1000.csv')
data.head()
```

```python
# Análisis descriptivo
data[["Revenue", "Profits", "Employees"]].describe()
```

```python
data.shape
```

```python
"""
Calcular la media, moda, mediana de cada una de las columnas númericas y explicar porque se muestra ese resultado para después
realizar un histograma
"""

print("Análisis descriptivo para cada columna numérica:")
for column in ["Revenue", "Profits", "Employees"]:
    media = data[column].mean()
    mediana = data[column].median()
    moda = data[column].mode()[0]
    print(f"{column} - Media: {media}, Mediana: {mediana}, Moda: {moda}")

print("\nDesviación estándar y varianza para cada columna numérica:")
for column in ["Revenue", "Profits", "Employees"]:
    desviacionestandar = data[column].std()
    varianza = data[column].var()
    print(f"{column} - Desviación estándar: {desviacionestandar}, Varianza: {varianza}")
```

```python
# Histograma para cada columna numérica
import matplotlib.pyplot as plt

for column in ["Revenue", "Profits", "Employees"]:
    plt.figure(figsize=(10, 6))
    plt.hist(data[column], bins=30, edgecolor='black')
    plt.title(f'Histograma de {column}')
    plt.xlabel(column)
    plt.ylabel('Frecuencia')
    plt.grid(axis='y', alpha=0.75)
    plt.show()
```

```python
# Explicacion de cada uno de los histogramas

"""
Columna Revenue: Se observa que la mayoría de las empresas tienen ingresos relativamente bajos, mientras que hay algunas empresas 
                 con ingresos extremadamente altos, lo que sugiere una distribución sesgada hacia la derecha.
Columna Profit: Similar a la columna de ingresos, la mayoría de las empresas tienen ganancias relativamente bajas, mientras que 
                hay algunas empresas con ganancias extremadamente altas, lo que también sugiere una distribución sesgada hacia 
                la derecha.
Columna Employees: Se muestra que la mayoría de las empresas tienen un número relativamente bajo de empleados, mientras que hay 
                 algunas empresas con un número extremadamente alto de empleados, lo que sugiere una distribución sesgada hacia 
                 la derecha.
"""
```

```python
# Boxplot para cada columna númerica

for column in ["Revenue", "Profits", "Employees"]:
    plt.figure(figsize=(10, 6))
    plt.boxplot(data[column])
    plt.title(f'Boxplot de {column}')
    plt.ylabel(column)
    plt.grid(axis='y', alpha=0.75)
    plt.show()
```
