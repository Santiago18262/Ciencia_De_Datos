```python
import math

class punto:

    def __init__(self, x = 0, y = 0):
        self.x = x
        self.y = y

    def __str__(self):
        return f"({self.x},{self.y})"
    
    def cuadrante (self):
        if (self.x == 0 and self.y == 0):       
            print(f"{self} pertenece al punto de origen")
        elif (self.x < 0 and self.y > 0):
            print(f"{self} pertenece al cuadrante 1")
        elif (self.x > 0 and self.y > 0):
            print(f"{self} pertenece al cuadrante 2")
        elif (self.x > 0 and self.y < 0):
            print(f"{self} pertenece al cuadrante 3")
        elif (self.x < 0 and self.y < 0):
            print(f"{self} pertenece al cuadrante 4")

    def vector(self, punto):
        x = punto.x - self.x
        y = punto.y - self.y
        return (f"Las coordenadas del vector es : ({x}, {y})")

    def distancia(self, punto):
        d = math.sqrt((punto.x - self.x)**2 + (punto.y - self.y)**2)
        print("La distancia entre los puntos {} y {} es {}".format(self,punto,d))

        
```

```python
p = punto(2,3)
r = punto(5,5)
p.vector(r)
p.distancia(r)
```
