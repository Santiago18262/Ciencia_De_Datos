```python
""" Ejemplo de implementacion con Programación Estructurada """

clientes = [
    {'Nombre': 'Hector ', 'Apellidos': 'Costa Guzman', 'dni':'11111111A'}, # Elemento 0
    {'Nombre': 'Juan ', 'Apellidos': 'Gonzalez Marquez', 'dni':'22222222B'} # Elemento 1
]

clientes
```

```python
def mostrar_clientes (ctes, dni):
    for c in ctes:
        if (dni == c['dni']):
            print('{}{}'.format(c['Nombre'],c['Apellidos']))
            return

    print('Cliente no encontrado')
```

```python
mostrar_clientes(clientes,'22222222B')
```

```python
def borrar_clientes(clientes,dni):
    for i,c in enumerate(clientes):
        if (dni == c['dni']):
            del(clientes[i])
            print(str(c), "> BORRADO")
            return
    print("Cliente no encontrado")
```

```python
borrar_clientes(clientes,'22222222B')
```

```python
def mostrar_todos(clientes):
    for c in clientes:
        print('{}{}'.format(c['Nombre'],c['Apellidos']))

mostrar_todos(clientes)
```

```python
def mostrar_todosID(clientes):
    for c in clientes:
        print('{} .- {}{}'.format(c['Id'],c['Nombre'],c['Apellidos']))

mostrar_todosID(clientes)
```

```python
class Pelicula:
    # Constructor de la clase
    def __init__(self, titulo, duracion, lanzamiento):
        self.titulo = titulo
        self.duracion = duracion
        self.lanzamiento = lanzamiento
        print("Se ha creado la pelicula", self.titulo)
    
    # Metodo destructor de la clase
    def __del__(self):
        print("Se ha borrado la pelicula", self.titulo)

    # Redefinimos el metodo str
    def __str__(self):
        return "{} Lanzado en {} con una duracion de {} minutos".format(self.titulo, self.lanzamiento, self.duracion)
    
    # Redefinimos el metodo len
    def __len__(self):
        return self.duracion
```

```python
class Catalogo:
    """
    Clase Catalogo: Para crear una coleccion de peliculas
    Input : Recibe peliculas
    Output: None
    Functions: 
                agregar: Esta funcion se utiliza para...
    """
    peliculas = []
    def __init__(self, peliculas = []):
        self.peliculas = peliculas
    
    def agregar(self, p):
        self.peliculas.append(p)

    def mostrar(self):
        for p in self.peliculas:
            print(p)
```

```python
p = Pelicula("El padrino", 175, 1972)
```

```python
cat = Catalogo([p])
```

```python
cat.mostrar()
```

```python
r = Pelicula("X MEN", 175, 1990)
```

```python
cat.agregar(r)
```

```python
cat.mostrar()
```
