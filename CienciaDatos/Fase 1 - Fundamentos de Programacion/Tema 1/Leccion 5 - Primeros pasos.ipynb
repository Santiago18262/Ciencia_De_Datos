```python
'''
La siguiente matriz (o lista con listas anidadas) debe cumplir una condicion, y es que en cada fila, el cuarto elemento
siempre debe ser el resultado de sumar los tres primeros. Â¿Eres capaz de modificar las sumas incorrectas utilizando la
tecnica de slicing?

Nota: Crear la matriz por el teclado

Ayuda: La funcion llamada sum(lista) devuelve la suma de todos los elementos de la lista
'''

# Declaracion de matriz
matriz = []

# Ingresar numeros a las listas por teclado
for i in range(3): # Para cada fila
    lista = [] 
    for j in range(3): # Para cada columna
        valor = int(input(f'Ingrese el valor para la posicion [{i}][{j}]: ')) 
        lista.append(valor) # Agrega el valor a la lista
    lista.append(sum(lista)) # Agrega el resultado de la suma de los 3 primeros digitos a la lista
    matriz.append(lista) # Agrega la lista a la matriz

# Imprimir la matriz
print(matriz)
```
