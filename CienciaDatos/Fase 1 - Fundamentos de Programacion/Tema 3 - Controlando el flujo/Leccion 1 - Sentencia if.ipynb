```python
'''
Realiza un programa que lea dos numeros por teclado y permita elegir 3 opciones de menú: 
- Mostrar una suma de los dos números
- Mostrar una resta de los dos números (el primero menos el segundo)
- Mostrar una multiplicación de los dos números
- En caso de no introducir una opcion valida, el programa informara de que no es correcta
'''

num1 = int(input("Introduce el primer numero: "))
num2 = int(input("Introduce el segundo numero: "))

print("Elige una opcion: ")

operador = input("Introduce el operador")
if operador == "suma": 
    print(num1 + num2)
elif operador == "resta": 
    print(num1 - num2)
else: 
    print("Introduce una opcion valida")

```
