```python
# Realiza un programa que sume todos los números enteros pares desde cero hasta el cien
n = 0
acumulador = 0 

while (n < 100):
    if(n % 2 == 0):
        acumulador += n
    n += 1
print(f"La suma total es: {acumulador}")
```

```python
# Realiza un programa que pida al usuario cuantos números quiere introducir. Luego lee 
# todos los números y realiza una media aritmetica
x = 0
valor = 0
numero = int(input("¿Cuántos números quieres introducir? "))
while x<numero:
    valor += float(input(f"Introduce un número: {x}"))
    x += 1
print(f"La media aritmetica es: {valor / numero}")
    
```

```python
print("Bienvenido al menu interactivo")
while(True):
    print("""Escriba una opcion
        1) Saludar
        2) Sumar 2 numeros
        3) Salir""")
    opcion = int (input("Ingresa un número: "))
    if opcion == 1:
        print ("Hola")
    elif opcion == 2:
        n1 = int(input("Ingresa el primer número: "))
        n2 = int(input("Ingresa el segundo número: "))
        print("El resultado de la suma: ", n1 + n2, "\n")
    elif opcion == 3:
        print("Hasta luego")
        break
    else:
        print("Opcion error")
```

```python
"""
Realiza un programa que lea un número impar por teclado. Si el usuario no introduce un número impar, debe
repetirse el proceso hasta que lo introduzca correctamente
"""

while(True):
    numero = int(input("Introduce un número impar: "))

    if numero % 2 != 0:
        print("El número es impar")
        break
    else: 
        print("Error")
```

```python
"""
Realiza un programa que pida al usuario un número entero del 0 al 9 y
que mientras el número no sea correcto se repita el proceso. Luego debe 
comprobar si el número se encuentra en la lista de números y notificarlo
"""
numero = 0
lista = [1, 2, 3, 4, 5]

while True:
    numero = int(input("Ingresa un número: "))
    if numero > 0 and numero < 10:
        if numero in lista:
            print("El numero esta en la lista")
            break
        else: 
            print("El numero no esta en la lista")
            break
    else: 
        print("Numero invalido")
```
