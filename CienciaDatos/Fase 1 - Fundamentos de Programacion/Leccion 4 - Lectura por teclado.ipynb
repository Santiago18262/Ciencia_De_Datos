```python
# Ejercicio con operaciones de cambio de tipo de dato
nombre = input("Captura el nombre: ")
apellido = input("Captura tu apellido: ")
edad = int(input("Captura tu edad: "))
numero_magico = float(input("Captura el numero magico: "))

cadena_final = nombre + " " + apellido + ": Tu numero de la suerte es el "
r = str(edad*numero_magico) # El str convierte a cadena de texto
cadena_final = cadena_final + r
```
