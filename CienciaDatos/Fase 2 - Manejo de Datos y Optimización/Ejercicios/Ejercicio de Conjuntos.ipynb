***1) Realiza un programa que siga las siguientes instrucciones***
* Pedir por teclado usuarios y administradores, preguntar si el usuario, si es asi, guardar en su respectivo conjunto
* Diseña un menú con 5 opciones: 1 capturar usuarios, 2 capturar administradores , 3 mostrar los datos, 4 eliminar usuario, 5 salir del programa
* Crea un conjunto llamado usuarios con los usuarios Marta, David, Elvira, Juan y Marcos
* Crea un conjunto llamado administradores con los administradores Juan y Marta
* Borre al administrador Juan del conjunto de administradores
* Añade a Marcos como nuevo administrador, pero no lo borres del conjunto de usuarios.
* Muestra todos los usuarios por pantalla de forma dinamica además debes de indicar cada uno de los usuarios es administrador o no

*Notas: Los conjuntos se pueden recorrer dinamicamente utilizando el bucle for de forma similar a una lista. Tambien cuentan con un metodo llamado .discard(elemento)
que sirve para borrar un elemento. Nota 2: Utiliza un ciclo while para capturar usuarios y administradores.*

```python
# Creamos los conjuntos 
usuarios = {'Marta', 'David', 'Elvira', 'Juan', 'Marcos'}
administradores = {'Juan', 'Marta'}

# Borramos a Juan de administradores usando discard
administradores.discard('Juan')

# Añade a Marcos como administrador
administradores.add('Marcos')

# Menu con opciones 
opcion = 0

while opcion != 5:
    print(''' 
----------MENU---------
1. Capturar Usuarios
2. Capturar Administradores
3. Mostrar los datos
4. Eliminar usuarios
5. Salir del programa''')

    opcion = input('\nIntroduce una opcion: ')
    if opcion == "1":
        print('\nEscribe salir para terminar de capturar')
        while True:
            nuevoUsuario = input('Nombre del usuario a agregar: ').capitalize() # .capitalize Convierte el primer carácter de la cadena a mayúscula.
            if nuevoUsuario.lower() == 'salir':
                break
            usuarios.add(nuevoUsuario)
    elif opcion == "2":
        print("\n(Escribe 'salir' para terminar de capturar)")
        while True:
            admin = input("Nombre del administrador a agregar: ").capitalize()
            if admin.lower() == "salir":
                break
            administradores.add(admin)
    elif opcion == "3":
        print("\n--- LISTADO DE USUARIOS ---")
        for u in usuarios:            # Verificamos si el usuario actual existe en el conjunto de administradores
            if u in administradores:
                print(f"- {u} (Administrador)")
            else:
                print(f"- {u} (Usuario estándar)")
    elif opcion == "4":
        eliminar = input("Nombre del usuario a eliminar: ").capitalize()
        usuarios.discard(eliminar)                # .discard() no lanza error si el elemento no existe
        administradores.discard(eliminar)         # También lo quitamos de admin por si acaso lo era
        print(f"\nIntento de eliminación de '{eliminar}' completado.")
    elif opcion == "5":
        print("\nSaliendo del programa...")
        break
    else:
        print("\nOpción no válida, intenta de nuevo.")
```
