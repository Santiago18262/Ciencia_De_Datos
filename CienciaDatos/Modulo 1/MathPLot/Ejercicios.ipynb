Pasos que debemos de considerar al gráficar 

* Importar 
* Crear datos
* Graficar
* Personalizar
* Mostrar o guardar

```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
# Datos
x = [10,20,30,40,50]
y = [22,40,60,80,100]

# Grafica 
plt.plot(x,y)

# Mostrar
plt.show()
```

```python
# El grafico debe contener lo siguiente

plt.plot(x,y) # Creamos la gráfica
plt.title("Mi Primera gráfica")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.grid(True)
plt.show()
```

```python
# Podemos cambiar estilos y marcadores en el grafico

plt.plot(x,y, color= "blue", linestyle= "--", marker = "o")
plt.title("Gráfica personalizada", color= "blue")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

```python
# Podemos visualizar varias lineas en una grafica

y2 = [10,20,30,40,50]
y = [16,30,40,50,60]

plt.plot(x, y, label = "Doble")
plt.plot(x, y2, label = "Normal")

plt.title("Dos lineas")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend() # Muestra una leyenda
plt.show()
```

```python
# Scatter plot

x = [1,2,3,4,5,6,7,11,12,13,14,15,16,18]
y = [10,12,14,15,16,17,21,22,23,24,25,26,27,28]
plt.scatter(x, y, color ='red')
plt.title("Gráfica de dispersion", color = "red")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# Gráfica de barras 

categorias = ["A", "B", "C", "D"]
valores = [10,15,7,12]

plt.bar(categorias, valores)
plt.title("Gráfica de barras")
plt.show()
```

```python
# Gráfica de pastel
plt.pie(valores, labels = categorias, autopct = '%1.3f%%')
plt.title("Gráfica de pastel")
plt.show()
```

```python
# Subgráficas

plt.subplot(2,2,2)
plt.plot(x,y)
plt.title("Linea")

plt.subplot(3,3,1)
plt.bar(categorias, valores)
plt.title("Barras")

plt.subplot(2,2,3)
plt.pie(valores, labels = categorias, autopct= "%1.3f%%")

plt.show()
```

```python
# Guardar la gráfica como imagen

plt.plot(x, y)
plt.title("Guardar Imagen")
plt.savefig("mi_gráfica.png")
plt.show()
```

```python
# Crear una lista con números del 1 al 20 en numpy
# Otra lista con temperaturas [20 temperaturas rango 10-50]
# Graficarlas
# Agregar titulo, etiquetas y color

numeros = np.arange(1, 21)  # Números del 1 al 20
temperaturas = np.random.randint(10, 51, 20)  # 20 temperaturas aleatorias entre 10 y 50

plt.bar(numeros, temperaturas, color='darkblue')
plt.title("Temperaturas")
plt.xlabel("Día")
plt.ylabel("Temperatura (°C)")
plt.show()
```
