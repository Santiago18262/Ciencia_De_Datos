```python
# Cargar el archivo y indexar por nombre y ordenar de forma permanente
import pandas as pd
df = pd.read_csv("employees.csv", index_col = "First Name", parse_dates = ["Start Date"])
df.sort_index(inplace = True)
df
```

```python
# Genera un reporte con filtros que contenga los registros de los empleados masculinos y de marketing
mask1 = df["Gender"] == "Male"
mask2 = df["Team"] == "Marketing"

df[mask1 & mask2]
```

```python
# Filtrar los empleados que son parte del Senior Managment o que empezaron antes de 1990
mask1 = df["Senior Management"] == True
mask2 = df["Start Date"] < "01-01-1990"

df[mask1 | mask2]
```

```python
# Filtra los empleados que se llaman Robert que son parte del Client Services o que empezaron despues de junio de 2016

mask1 = df.index == "Robert"
mask2 = df["Team"] == "Client Services"
mask3 = df["Start Date"] > "30-06-2016"

df[mask1 & (mask2 | mask3)]
```

```python
# Filtra los empleados que son parte del Legal, Sales o Product

mask1 = df["Team"] == "Legal" 
mask2 = df["Team"] == "Sales"
mask3 = df["Team"] == "Product"

df[mask1 | mask2 | mask3]
```

```python
mask = df["Team"].isin(["Legal", "Sales", "Product"])
df[mask]
```

```python
mask1 = (df["Team"] == "Legal") | (df["Team"] == "Sales") | (df["Team"] == "Product")

df[mask1] 
```

```python
# Filtra los empleados que no tienen un  equipo asignado (Team es null)

df["Team"].isnull()
```

```python
# Crea un reporte que contenga los empleados que ganan entre 60000 y 70000 dolares
mask = df["Salary"].between(60000, 70000)

df[mask]
```

```python
# Nomina de todos los empleados de marketing 

mask = df["Team"] == "Marketing"
nomina = df.loc[mask, "Salary"].sum()

print(f"El total de la nomina es: ${nomina} dólares")
```

```python
# Todos los empleados que empezaron a trabajar entre el 1 de enero de 1991 y el 1 de enero de 1992

mask = df["Start Date"].between("01-01-1991", "01-01-1992")

df[mask]
```

```python
# Cuántos empleado unicos hay en el dataframe

df.index.is_unique

df.index.nunique()
```

```python
# Cuantos equipos son los equipos (Team) unicos en el dataframe
num_equipos = df['Team'].nunique()
print(num_equipos)
```

```python
# Cuantos empleados únicos hay en el dataframe que no tienen un equipo asignado
empleados_sin_equipo = df[df["Team"].isnull()]
cantidad_unicos = empleados_sin_equipo.index.nunique()
print(cantidad_unicos)
```
