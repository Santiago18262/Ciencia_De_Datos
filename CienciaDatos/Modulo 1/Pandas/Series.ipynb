## Método .apply() 

```python
google.head(10)
```

```python
def classify_performance(num):
    if num < 300:
        return "OK"
    elif num >= 300 and num < 650:
        return "Satisfactory"
    else:
        return "Incredible"
```

```python
google.apply(classify_performance)
```

```python
google.apply(classify_performance).tail()
```

```python
google.apply(lambda stock_p : stock_p/2)

# def stock_P:
# stock_p/2
```
