```python
import pandas as pd

# 1. Cargar datos
df = pd.read_csv("employees.csv")

# 2. Asegurar que la columna sea numérica (por si acaso hay basura o texto)
df["Salary"] = pd.to_numeric(df["Salary"], errors='coerce')

# 3. Aplicar el rango correcto: 60,000 a 70,000
mask = df["Salary"].between(60000, 70000)

# 4. Guardar el resultado en un nuevo DataFrame y mostrarlo
df_filtrado = df[mask]

# Ver los resultados
print(df_filtrado)
```
