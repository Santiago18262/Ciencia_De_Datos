## Drop rows eith Null Values ##

```python
nba = pd.read_csv("nba.csv")
```

```python
nba.tail(3)
```

```python
nba.dropna # Borra todos los valores null
```

```python
nba.dropna(how = "all") # Borra solo las filas que tienen todos los valores null
```

```python
nba.dropna(how = "all", inplace = True) # Borra las filas que tienen todos los valores null y lo guarda en el mismo DataFrame
```

```python
nba.dropna(axis = 1) # Borra las columnas que tienen al menos un valor null, axis = 1 se refiere a columnas, axis = 0 se refiere a filas
```

```python
nba.dropna(subset = ["Salary"]) # Borra las filas que tienen valores null en la columna Salary
```

```python
# Ingresar un jugador por teclado y localizarlo en el dataFrame con nombre y apellido

import pandas as pd

# 1. Cargar el archivo CSV
df = pd.read_csv('nba.csv')

# 2. Solicitar el nombre por teclado
nombre_buscado = input("Ingresa el nombre y apellido del jugador: ")

# 3. Localizar al jugador
# Usamos .str.lower() en ambos lados para que la búsqueda sea más flexible
resultado = df[df['Name'].str.lower() == nombre_buscado.lower()]

# 4. Mostrar el resultado
if not resultado.empty:
    print("\n--- Jugador Localizado ---")
    print(resultado)
else:
    print(f"\nNo se encontró a '{nombre_buscado}' en el archivo nba.csv.")
```
