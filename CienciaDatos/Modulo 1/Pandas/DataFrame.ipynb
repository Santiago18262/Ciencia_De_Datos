```python
# Ingresar un jugador por teclado y localizarlo en el dataFrame con nombre y apellido

import pandas as pd

# 1. Cargar el archivo CSV
df = pd.read_csv('nba.csv')

# 2. Solicitar el nombre por teclado
nombre_buscado = input("Ingresa el nombre y apellido del jugador: ")

# 3. Localizar al jugador
# Usamos .str.lower() en ambos lados para que la búsqueda sea más flexible
resultado = df[df['Name'].str.lower() == nombre_buscado.lower()]

# 4. Mostrar el resultado
if not resultado.empty:
    print("\n--- Jugador Localizado ---")
    print(resultado)
else:
    print(f"\nNo se encontró a '{nombre_buscado}' en el archivo nba.csv.")
```
