```python
class AnalizadorTexto:
    
        cadena = input("Introduce un texto: ")
        cadena = cadena.lower()
        palabras = cadena.split()
        frecuencia = {}

        for palabra in palabras:
            if palabra in frecuencia:
                frecuencia[palabra] += 1
            else:
                frecuencia[palabra] = 1

        # Mostrar resultados
        print("\nFrecuencia de palabras: ")

        for palabra in frecuencia:
            print(palabra, ":", frecuencia[palabra])

        # Mostrar las 5 palabras mas repetidas
        print("\nLas 5 palabras más repetidas: ")
        top_5 = sorted(frecuencia, reverse = True)[:5]
        print(top_5)

        palabras_ordenadas = sorted(frecuencia, reverse = True)
        for palabra in palabras_ordenadas[:5]:
            print(palabra, ":", frecuencia[palabra])

        ## Tasa de aparicion
        total_palabras = len(palabras)
        print("\nTasa de aparición de cada palabra: ")
        for palabra in frecuencia:
            tasa = frecuencia[palabra] / total_palabras
            print(palabra, ":", tasa)
```
