```python
class Punto:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        return "({}, {})".format(self.x, self.y)
```

```python
p1 = Punto(5, 6)
p1.__str__()
```

```python
class Rectangulo:
    def __init__(self, p1, p2):
        self.p1 = p1
        self.p2 = p2
    
    def area(self):
        base = abs(self.p2.x - self.p1.x)
        altura = abs(self.p2.y - self.p1.y)
        return base * altura
```

```python
esquina_a = Punto(2,2)
esquina_b = Punto(10,5)

mi_rectangulo = Rectangulo(esquina_a, esquina_b)

print(f"Punto 1: {mi_rectangulo.p1}")
print(f"Punto 2: {mi_rectangulo.p2}")
print(f"El área es: {mi_rectangulo.area()}") # (10-2) * (5-2) = 8 * 3 = 24
```
