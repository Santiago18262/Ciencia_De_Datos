<h1>Acceso a datos</h1>
<h4>La notacion de matriz y los rangos se pueden usar para acceder de manera eficiente</h4>

```python
#Hacer una matriz 2x3 e imprimir la primera y ultima lista
# Luego imprimir un elemento concreto y los valores de la columna 3

import numpy as np
mylist = [[1,2,3],[4,5,6]] #Matriz 2x3
myarray = np.array(mylist) # convertimos la matriz a NP
print(myarray) #imprimimos la matriz por pantalla
```

```python
print(myarray.shape)
print(f"First row: {myarray[0]}")
print(f"Last row: {myarray[-1]}")

print(f"Specific row and col: {myarray[1,-1]}")

print(f"Whole col: {myarray[0:2,1:2]}")
```

<h1>Operadores aritmeticos</h1>

```python
# Crear dos vectores y sumarlos directamente en aritmetica
import numpy as np # Ya esta cargada previamente.

#Creamos 2 listas y la transformamos a NP
myarray1= np.array([2, 2, 2, 1])
myarray2= np.array([3,3,3,8])

print(f"Addition: {myarray1 +myarray2}")
print(f"Multiplicacion: {myarray1*myarray2}")

print(f"Addition + Multiplicacion: {(myarray1 +myarray2)+(myarray1*myarray2)}")
```

<h2>Graficos lineales</h2>

```python
#Crear un vector y representar un grafico linea con sus valores
import matplotlib.pyplot as plt
import numpy as np
myarray = np.array([1,2,3,4,5,6,7])#Creamos un arreglo NP
plt.plot(myarray)#Creamos el objeto Grafica
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show
```

```python
#Hcer dos vectors y representar los puntos en un grafico de dispersion
import matplotlib.pyplot as plt
import numpy as np
x= np.array([1,2,3])
y = np.array([2,4,6])
plt.scatter(x,y)
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show
```

<h1>Pandas</h1>

<h1>Dataframe</h1>

```python
#Crear un Dataframe e imprimirlo
import pandas as pd
import numpy as np

myarray = np.array([[1,2,3], [4,5,6],[34,22,13], [34,45,66]])
rownames = ['a','b','c','d']
colnames = ['one','two','three']
df_my = pd.DataFrame(myarray,index=rownames,columns=colnames)
df_my
```

```python
#df_my.head(1) #Los primeros elementos del data frame
df_my.tail(2) #Los ultimos elementos del data frame
```

```python
df_my.shape # Dimensiones del data frame
df_my.describe # Estructura del data frame sin estilo
```
